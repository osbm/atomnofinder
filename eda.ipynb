```python
def calculate_score(df_rdkit, dr_re):
    score = np.prod(df_rdkit.shape)

    for rdkit_row, re_row in zip(df_rdkit.iterrows(), dr_re.iterrows()):
        rdkit_atom_nos = set(rdkit_row[1][1:])  # Assuming the atom number columns start from index 1
        re_atom_nos = set(re_row[1][1:])

        if rdkit_atom_nos == re_atom_nos:
            score -= 1

    return score

calculate_score(df_rdkit, df_re)
```
